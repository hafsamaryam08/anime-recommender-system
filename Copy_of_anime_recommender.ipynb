```python
def convert_genres(text):
    if isinstance(text, str):
        return [genre.strip() for genre in text.split(",")]
    return text

anime["Genres"] = anime["Genres"].apply(convert_genres)

print(anime["Genres"].head())
```

```python
def convert(text):
    if isinstance(text, str):
       return text.split()
    return text

anime["Description"] = anime["Description"].apply(convert)
anime["synopsis"] = anime["synopsis"].apply(convert)

print(anime[["Description", "synopsis"]].head())
```

```python
anime.head()
```

```python
def convert(text):
    if isinstance(text, str):
        return [word.strip() for word in text.split() if word]
    return text

anime["Description"] = anime["Description"].apply(convert)
anime["synopsis"] = anime["synopsis"].apply(convert)

print(anime[["Description", "synopsis"]].head())
```

```python
anime["tags"] = anime["Genres"] + anime["Description"] + anime["synopsis"]
```

```python
anime.head()
```

```python
new_df = anime[["Name","Rating","tags"]]
```

```python
new_df
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(anime_name):
    if anime_name not in new_df['Name'].values:
        print(f"'{anime_name}' not found in the dataset. Please choose a valid anime.")
        return []
    anime_index = new_df[new_df['Name'] == anime_name].index[0]
    distances = similarity[anime_index]
    anime_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    recommendations = [new_df.iloc[i[0]].Name for i in anime_list]
    return recommendations
```

```python
recommend('Fullmetal Alchemist: Brotherhood')
```
